# Why the SVD Rules

SVD = Singular Value Decomposition

$$
M = U D V^{T}
$$


The SVD is the archetypal matrix factorization

```python
srand(1)
A = rand(5, 5)
```

```python
U, d, V = svd(A)
```

```python
D = diagm(d)
```

```python
U * D * V'
```

```python
A
```

Why should you care that you can write 1 matrix in terms of 3 matrices?

Because the SVD is the perfect summary of a matrix

## Social Network Analysis

Let's show how we can use the SVD to make sense of social networks

```python
monks = readcsv("sna/monks.csv", Int)
```

```python
U, d, V = svd(monks)
```

```python
X = U * diagm(sqrt(d))
```

```python
groups = vcat(repeat([1], inner = [7]),
	          repeat([2], inner = [7]),
	          repeat([3], inner = [4]))
```

```python
plot(x = X[:, 1], y = X[:, 2], group = groups, kind = :scatter)
```

![Sampson's Monks](files/images/Fig2.png)

The U matrix from the SVD gives us one representation of the social graph

Entries that are similar in U are also similar in the social network

So we can use the columns of U to visualize a social network

## Text Analysis

```python
dir = "sotu"
```

```python
documents =
  [lowercase(readall(joinpath(dir, path))) for path in readdir(dir)];
```

```python
tokenize(doc::String) = matchall(r"\w+", doc)
```

```python
tokens = [tokenize(doc) for doc in documents];
```

```python
all_tokens = Set()
for tokenized_doc in tokens
    for term in tokenized_doc
        add!(all_tokens, term)
    end
end
```

```python
index = Dict(collect(all_tokens), 1:length(all_tokens))
```

```python
function makedtm(tokens::Vector, index::Dict)
    d = length(tokens)
    L = length(all_tokens)
    m = zeros(Int, d, L)
    for i in 1:d
        for token in tokens[i]
            j = index[token]
            m[i, j] += 1
        end
    end
    return m
end
```

```python
dtm = makedtm(tokens, index)
```

```python
U, d, V = svd(dtm)
```

```python
X = U * diagm(sqrt(d))
```

```python
gr = vcat(repeat([1], inner = [50]),
          repeat([2], inner = [50]),
          repeat([3], inner = [50]),
          repeat([4], inner = [50]),
          repeat([5], inner = [14]))
```

```python
plot(x = X[:, 1], y = X[:, 2], group = gr, kind = :scatter)
```

![Representation 1 of Text Data](files/images/Fig3.png)

```python
X = dtm * V
```

```python
plot(x = X[:, 1], y = X[:, 2], group = gr, kind = :scatter)
```

![Representation 2 of Text Data](files/images/Fig4.png)

What was this second approach?

We rotated the original data by the matrix V

This rotation forms the principal components of the original data

Let's see the rotation on 2D data to get a sense for it

```python
using Distributions
```

```python
μ = [0.0, 0.0]
```

```python
Σ = [1.0 0.9;
     0.9 1.0]
```

```python
distr = MultivariateNormal(μ, Σ)
```

```python
X = rand(distr, 1_000)
```

```python
plot(x = vec(X[1, :]), y = vec(X[2, :]), kind = :scatter)
```

![Raw Data](files/images/Fig5.png)

```python
U, d, V = svd(X)
```

```python
D = diagm(d)
```

```python
Xr = U * X
```

```python
plot(x = vec(Xr[1, :]), y = vec(Xr[2, :]), kind = :scatter)
```

![Rotated Data](files/images/Fig6.png)

Principal components analysis (PCA) just involves rotating the data

The SVD gives us the matrix we need to perform this rotation

## Building a Market Index

Let's make a simple market index like the DJI using the SVD

```python
using DataFrames
```

```python
df = readtable("stocks/stocks.csv")
```

```python
dji = readtable("stocks/dji.csv")
```

```python
prices = matrix(df[:, 2:end])
```

```python
dji = vector(dji[1])
```

```python
U, d, V = svd(prices)
```

```python
X = prices * V
```

```python
mydji = (X[:, 1] - mean(X[:, 1])) ./ std(X[:, 1])
```

```python
plot(x = repeat([1:length(dji)], outer = [2]),
	 y = vcat(mydji, dji),
	 group = repeat(["SVD", "DJI"], inner = [length(dji)]),
	 kind = :line)
```

![Market Index: Take 1](files/images/Fig7.png)

```python
scaled_dji = (dji - mean(dji)) ./ std(dji)
```

```python
plot(x = repeat([1:length(dji)], outer = [2]),
	 y = vcat(mydji, scaled_dji),
	 group = repeat(["SVD", "DJI"], inner = [length(dji)]),
	 kind = :line)
```

![Market Index: Take 2](files/images/Fig8.png)

```python
plot(x = repeat([1:length(dji)], outer = [2]),
	 y = vcat(-mydji, scaled_dji),
	 group = repeat(["SVD", "DJI"], inner = [length(dji)]),
	 kind = :line)
```

![Market Index: Take 3](files/images/Fig9.png)

## Voting Records

```python
votes = readtable("votes/senate112.csv")
```

```python
dm = DataArray(votes[:, 1:424])
```

```python
senator_names = vector(votes[:, 425])
```

```python
republicans = int([ismatch(r"\(R ", name) for name in senator_names])
```

```python
U, d, V = svd(datafloat(dm), 2)
```

```python
X = U * diagm(d)
```

```python
plot(x = X[:, 1], y = X[:, 2], group = republicans, kind = :scatter)
```

![Representation of Vote Data](files/images/Fig10.png)
